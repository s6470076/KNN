```python
import pandas as pd
```

## 주소를 이용한 데이터셋 로드

### 1.주소 저장하기

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
```

```python
url
```

### 2.컬럼 셋팅

```python
names = ['sepal-length','sepal-width','setal-length','setal-width','Class']
```

### 3.iris데이터셋 읽어오기

```python
dataset = pd.read_csv(url,names=names)
```

```python
dataset
```

## target값 나누기

```python
x = dataset.iloc[:,:-1].values # Class를 제외한 값을 저장한다
y = dataset.iloc[:,4].values   # Class값만 저장한다
```

```python
x, x.shape
```

```python
y, y.shape
```

## 모델링

### 1.트레인데이터와 테스트데이터 분리하기

```python
from sklearn.model_selection import train_test_split

# 트레인데이터 : 테스트데이터 == 8:2
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2) # 테스트사이즈==0.2
```

```python
x_train, x_train.shape
```

### 2. 데이터 변환(데이터 정규화)
- 훈련시킬 수 있는 형태로 변환

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x_train)

x_train = scaler.transform(x_train)
x_test  = scaler.transform(x_test) 
```

```python
x_train
```

### 3.모델 제작

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier = KNeighborsClassifier(n_neighbors=5)
```

```python
classifier.fit(x_train,y_train)
```

### 예측하기

```python
y_pred = classifier.predict(x_test)
y_pred
```

### 평가하기

```python
from sklearn.metrics import classification_report, confusion_matrix

#혼돈행렬로 성능테스트
print(confusion_matrix(y_test,y_pred))#실제값과 예측값을 비교해서 혼돈행렬
                                      #대각선으로 숫자가 나올 경우 우수한 모델
print(classification_report(y_test,y_pred))
```
